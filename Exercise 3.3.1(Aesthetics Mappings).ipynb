**3. Map a continuous variable to color, size, and shape. How do these aesthetics behave differently for categorical versus continuous variables?**

```r
ggplot(data = mpg)+
    geom_point(mapping = aes(x = displ, y = hwy, color = cty))
```

_Ggplot when mapping a continous variable with color aesthetic creates a scale that varies from light(here light blue) to dark(here dark blue). The data points get darker as the city miles per gallon increments and lighter as it decrements._

```r
ggplot(data = mpg)+
    geom_point(mapping = aes(x = displ, y = hwy, size = cyl))
```

_Ggplot when mapping a continous variable with size aesthetic creates a scale of circles that varies from small to large. The data points get larger as the number of cylinders increments and smaller as it decrements._

```r
ggplot(data = mpg)+
    geom_point(mapping = aes(x = displ, y = hwy, shape = year))
```

_Color can vary on a scale so can size and thus can represent varying continous variables in scale. Shapestake distinct unvarying forms thus are categorical and thus can not represent continous variables and but represent categorical variables. Hence the error._

*4. What happens if you map the same variable to multiple aesthetics?**

```r
ggplot(data = mpg)+
    geom_point(mapping = aes(x = displ, y = hwy, color = displ, size = displ))
```

_So when you map the same continous variable all the aesthetics argued are mapped and different legends made for them._

```r
ggplot(data = mpg)+
    geom_point(mapping = aes(x = displ, y = hwy, color = fl, shape = fl))
```

_So when you map the same continous variable all the aesthetics argued are mapped and a single legend made for them._

**5. What does the stroke aesthetic do? What shapes does it work with? (Hint: use ?geom_point.)**

```r
ggplot(data = mpg) +
  geom_point(mapping = aes(x = displ, y = hwy), shape = 21,
             fill = 'red', size = 5, stroke = 5, color = 'white')
```
