**1. What geom would you use to draw a line chart? A boxplot? A histogram? An area chart?**

_Line plot_

```r
ggplot(data = mpg)+
    geom_line(mapping = aes(x = displ, hwy))
```

```r
ggplot(data = mpg)+
    geom_boxplot(mapping = aes(x = class, hwy))
```

```r
ggplot(data = mpg)+
    geom_histogram(mapping = aes(x = hwy))
```

```r
ggplot(data = mpg)+
    geom_area(mapping = aes(x = displ, y = hwy))
```

**3. What does show.legend = FALSE do? What happens if you remove it? Why do you think I used it earlier in the chapter?**

_``show.legend = FALSE`` does not show the legend that is rendered when the plot is made. If you remove it since by default ``show.legend = TRUE`` a legend will be rendered. The reason it was used earlier by the below code is because it will just be substantiating the obvious thus not needed._  

```r
ggplot(data = mpg) +
    geom_smooth(
    mapping = aes(x = displ, y = hwy, color = drv),
    show.legend = FALSE)
```

**4. What does the se argument to geom_smooth() do?**

The ``se`` argument in ``geom_smooth()`` determines whether the confidence interval around the smooth lines is shown.

**5. Will these two graphs look different? Why/why not?**

```r
ggplot(data = mpg, mapping = aes(x = displ, y = hwy)) +
    geom_point() +
    geom_smooth()
ggplot() +
    geom_point(data = mpg, mapping = aes(x = displ, y = hwy)) +
    geom_smooth(data = mpg, mapping = aes(x = displ, y = hwy))
```

_Nah they do not look different, they look the same. Why: They both are mapping the same aesthetics on the same variables from the same data. The first one globally the second one locally._
