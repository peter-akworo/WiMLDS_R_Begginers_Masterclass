**1. What is the problem with this plot? How could you improve it?**

```r
ggplot(data = mpg, mapping = aes(x = cty, y = hwy)) +
    geom_point()
```

_There is overlapping of data points because there are multiple observations for value combinations of hwy and cty_ 

_We can improve the plot using a jitter position adjustment_

```r
ggplot(data = mpg, mapping = aes(x = cty, y = hwy)) +
    geom_point(position = "jitter")
```

*Or use the `geom_jitter()` instead of `geom_point()`*

```r
ggplot(data = mpg, mapping = aes(x = cty, y = hwy)) +
    geom_jitter()
```

**2. What parameters to geom_jitter() control the amount of jittering?**

```r
?geom_jitter()
```

*The parameters `width` and `height` control the amount of jittering in a `geom_jitter()` plot*

```r
ggplot(data = mpg, mapping = aes(x = cty, y = hwy)) +
    geom_jitter(width = 0, height = 0)
```

_On default_ 

_`Width` and `height` parameters set_ 

```r
ggplot(data = mpg, mapping = aes(x = cty, y = hwy)) +
    geom_jitter(width = 5, height = 10)
```

**3. Compare and contrast geom_jitter() with geom_count()**

_`geom_jitter()` randomly variates the data points across the plot to avoid overlapping. This comes at a cost of statistical accuracy.
`geom_count()` sizes the data points relative to the frequency of the observations_ 

```r
ggplot(data = mpg, mapping = aes(x = cty, y = hwy)) +
    geom_jitter()
```

```r
ggplot(data = mpg, mapping = aes(x = cty, y = hwy)) +
    geom_count()
```
